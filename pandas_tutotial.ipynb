`df.fare.map(lambda p: p - df_fare_mean)` dla każdej wartości p w serii fare wywołuje funkcję, która zwraca różnicę między tą wartością a uprzednio obliczoną średnią, czyli tworzy nową serię odchyleń od wartości średnie

```python
df_fare_mean = df.fare.mean()
print(df.fare.map(lambda p: p - df_fare_mean))
print('\n')     # to samo
print(df.fare - df_fare_mean)
```

```python
def remean_points(row):
    row.fare = row.fare - df_fare_mean
    return row

df.apply(remean_points, axis='columns')
```

Znalezienie indeksu maksymalnego stosunku `fare/age`

```python
df_clean = df.dropna(subset=['fare', 'age'])
df_idx = (df_clean.fare / df_clean.age).idxmax()
df_passenger = df.loc[df_idx, ['who', 'fare', 'age']]

print(f"Najwyższy wskaźnik fare/age ma pasażer o indeksie {df_idx}:")
print(f"  Typ: {df_passenger['who']}, Fare: {df_passenger['fare']}, Age: {df_passenger['age']}")
```
